```python
# importando pacotes
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

# Modelo Bernoulli

```python
# gerando amostras para distribuições bernoulli

amostra_bernoulli_p_03 = np.random.binomial(1, 0.3, 1000)
amostra_bernoulli_p_05 = np.random.binomial(1, 0.5, 1000)
amostra_bernoulli_p_07 = np.random.binomial(1, 0.7, 1000)

Bernoulli_df = pd.DataFrame({
    'Bernoulli p=0.3': amostra_bernoulli_p_03,
    'Bernoulli p=0.5': amostra_bernoulli_p_05,
    'Bernoulli p=0.7': amostra_bernoulli_p_07
})
```

```python
# Criar uma figura para os subplots
plt.figure(figsize=(15, 5))

# Iterar sobre as colunas do DataFrame e criar histogramas
for i, col in enumerate(Bernoulli_df.columns):
    plt.subplot(1, 3, i+1)
    sns.histplot(Bernoulli_df[col], bins=8, kde=False, color=sns.color_palette()[i], stat='probability')
    plt.title(col)
    plt.xlabel('Valores')
    plt.ylabel('Frequência Relativa')

# Exibir o gráfico
plt.show()
```

# Modelo Binomial

```python
# gerando amostras para distribuição binomial com n = 5

amostra_binomial_n_5_p_03 = np.random.binomial(5, 0.3, 1000)
amostra_binomial_n_5_p_05 = np.random.binomial(5, 0.5, 1000)
amostra_binomial_n_5_p_07 = np.random.binomial(5, 0.7, 1000)

Binomial_n_5_df = pd.DataFrame({
    'Binomial n=5 p=0.3': amostra_binomial_n_5_p_03,
    'Binomial n=5 p=0.5': amostra_binomial_n_5_p_05,
    'Binomial n=5 p=0.7': amostra_binomial_n_5_p_07
})
```

```python
# gerando amostras para distribuição binomial com n = 10

amostra_binomial_n_10_p_03 = np.random.binomial(10, 0.3, 1000)
amostra_binomial_n_10_p_05 = np.random.binomial(10, 0.5, 1000)
amostra_binomial_n_10_p_07 = np.random.binomial(10, 0.7, 1000)

Binomial_n_10_df = pd.DataFrame({
    'Binomial n=10 p=0.3': amostra_binomial_n_10_p_03,
    'Binomial n=10 p=0.5': amostra_binomial_n_10_p_05,
    'Binomial n=10 p=0.7': amostra_binomial_n_10_p_07
})
```

```python
# Criar uma figura para os subplots
plt.figure(figsize=(15, 15))

# Iterar sobre as colunas do DataFrame e criar histogramas da primeira linha
for i, col in enumerate(Binomial_n_5_df.columns):
    plt.subplot(2, 3, i+1)
    sns.histplot(Binomial_n_5_df[col], bins=8, kde=False, color=sns.color_palette()[i], stat='probability')
    plt.title(col)
    plt.xlabel('Valores')
    plt.ylabel('Frequência Relativa')

for i, col in enumerate(Binomial_n_10_df.columns):
    plt.subplot(2, 3, i+4)
    sns.histplot(Binomial_n_10_df[col], bins=8, kde=False, color=sns.color_palette()[3+i], stat='probability')
    plt.title(col)
    plt.xlabel('Valores')
    plt.ylabel('Frequência Relativa')


# Exibir o gráfico
plt.show()
```

# Modelo Geométrico

```python
# gerando amostras para distribuição geometrica

amostra_geometrica_p_03 = np.random.geometric(0.3, 1000)
amostra_geometrica_p_05 = np.random.geometric(0.5, 1000)
amostra_geometrica_p_07 = np.random.geometric(0.7, 1000)

Geometrica_df = pd.DataFrame({
    'Geometrica p=0.3': amostra_geometrica_p_03,
    'Geometrica p=0.5': amostra_geometrica_p_05,
    'Geometrica p=0.7': amostra_geometrica_p_07
})
```

```python
# Criar uma figura para os subplots
plt.figure(figsize=(15, 5))

# Iterar sobre as colunas do DataFrame e criar histogramas
for i, col in enumerate(Geometrica_df.columns):
    plt.subplot(1, 3, i+1)
    sns.histplot(Geometrica_df[col], bins=20, kde=False, color=sns.color_palette()[i], stat='probability')
    plt.title(col)
    plt.xlabel('Valores')
    plt.ylabel('Frequência Relativa')

# Exibir o gráfico
plt.show()
```

# Modelo Poisson

```python
# gerando amostras para distribuição poisson

amostra_poisson_lmbd_3 = np.random.poisson(3, 1000)
amostra_poisson_lmbd_5 = np.random.poisson(5, 1000)
amostra_poisson_lmbd_7 = np.random.poisson(7, 1000)

Poisson_df = pd.DataFrame({
    'Poisson λ=3': amostra_poisson_lmbd_3,
    'Poisson λ=5': amostra_poisson_lmbd_5,
    'Poisson λ=7': amostra_poisson_lmbd_7
})
```

```python
 # Criar uma figura para os subplots
plt.figure(figsize=(15, 5))

# Iterar sobre as colunas do DataFrame e criar histogramas
for i, col in enumerate(Poisson_df.columns):
    plt.subplot(1, 3, i+1)
    sns.histplot(Poisson_df[col], bins=20, kde=False, color=sns.color_palette()[i], stat='probability')
    plt.title(col)
    plt.xlabel('Valores')
    plt.ylabel('Frequência Relativa')

# Exibir o gráfico
plt.show()
```

```python
distribuicoes_df = pd.concat([Bernoulli_df,Binomial_n_5_df, Binomial_n_10_df, Geometrica_df, Poisson_df],axis = 1)
```

```python
rows = 5
cols = 3

plt.figure(figsize=(15, 10))

for i, col in enumerate(distribuicoes_df.columns):
    plt.subplot(rows, cols, i+1)  # Organizar os subplots em um grid 5x3
    sns.histplot(distribuicoes_df[col], kde=False, color=sns.color_palette()[i % 10], stat='probability')
    plt.title(col)  # Usar o nome da coluna como título
    plt.xlabel('Valores')
    plt.ylabel('Frequência Relativa')

plt.tight_layout()

# Exibir o gráfico
plt.show()
```

# Modelo uniforme

```python
rv_uniform01 = stats.uniform(0, 1)
rv_uniform24 = stats.uniform(2, 2)
rv_uniform39 = stats.uniform(3, 6)

# plotando a função de densidade de cada uma das variaveis aleatorias
xx = np.linspace(-10, 10, 1000)

plt.plot(xx, rv_uniform01.pdf(xx), label='Uniform(0,1)')
plt.plot(xx, rv_uniform24.pdf(xx), label='Uniform(2,4)')
plt.plot(xx, rv_uniform39.pdf(xx), label='Uniform(3,9)')
plt.legend()
plt.show()
```

# Modelo Exponencial

```python
rv_exponencial1 = stats.expon(scale=1)
rv_exponencial2 = stats.expon(scale=2)
rv_exponencial3 = stats.expon(scale=3)

# plotando a função de densidade de cada uma das variaveis aleatorias

fig2 = plt.figure()
xxp = np.linspace(0, 10, 1000)
plt.plot(xxp, rv_exponencial1.pdf(xxp), label='Exp(1)')
plt.plot(xxp, rv_exponencial2.pdf(xxp), label='Exp(2)')
plt.plot(xxp, rv_exponencial3.pdf(xxp), label='Exp(3)')
plt.legend()
plt.show()
```

# Modelo Normal

```python
rv_normal01 = stats.norm(0, 1)
rv_normal21 = stats.norm(2, 1)
rv_normal31 = stats.norm(3, 1)

# plotando a função de densidade de cada uma das variaveis aleatorias

fig3 = plt.figure()
plt.plot(xx, rv_normal01.pdf(xx), label='Normal(0,1)')
plt.plot(xx, rv_normal21.pdf(xx), label='Normal(2,1)')
plt.plot(xx, rv_normal31.pdf(xx), label='Normal(3,1)')
plt.legend()
plt.show()
```

```python
rv_normal01 = stats.norm(0, 1)
rv_normal02 = stats.norm(0, 2)
rv_normal03 = stats.norm(0, 3)

# plotando a função de densidade de cada uma das variaveis aleatorias

fig4 = plt.figure()
plt.plot(xx, rv_normal01.pdf(xx), label='Normal(0,1)')
plt.plot(xx, rv_normal02.pdf(xx), label='Normal(0,2)')
plt.plot(xx, rv_normal03.pdf(xx), label='Normal(0,3)')
plt.legend()
plt.show()
```

```python
# Plotando as quatro figuras juntas como subplots

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0,0].plot(xx, rv_uniform01.pdf(xx), label='Uniform(0,1)')
axs[0,0].plot(xx, rv_uniform24.pdf(xx), label='Uniform(2,4)')
axs[0,0].plot(xx, rv_uniform39.pdf(xx), label='Uniform(3,9)')
axs[0,0].legend()
axs[0,0].set_ylabel('f(x)',rotation=0)

axs[0,1].plot(xxp, rv_exponencial1.pdf(xxp), label='Exp(1)')
axs[0,1].plot(xxp, rv_exponencial2.pdf(xxp), label='Exp(2)')
axs[0,1].plot(xxp, rv_exponencial3.pdf(xxp), label='Exp(3)')
axs[0,1].legend()
axs[0,1].set_ylabel('f(x)',rotation=0)

axs[1,0].plot(xx, rv_normal01.pdf(xx), label='Normal(0,1)')
axs[1,0].plot(xx, rv_normal21.pdf(xx), label='Normal(2,1)')
axs[1,0].plot(xx, rv_normal31.pdf(xx), label='Normal(3,1)')
axs[1,0].legend()
axs[1,0].set_ylabel('f(x)',rotation=0)

axs[1,1].plot(xx, rv_normal01.pdf(xx), label='Normal(0,1)')
axs[1,1].plot(xx, rv_normal02.pdf(xx), label='Normal(0,2)')
axs[1,1].plot(xx, rv_normal03.pdf(xx), label='Normal(0,3)')
axs[1,1].legend()
axs[1,1].set_ylabel('f(x)',rotation=0 )

plt.show()
```
